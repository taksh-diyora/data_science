# Central Limit Theorem:

- ##### CLT states that when plotting a sample distribution of means, the means of the sample will be equal to the population mean and the sample distribution will approach normal distribution with variance equal to standard error.

#### Assumptions behind the CLT:

- ##### Sample data must be sampled and selected randomly from the population
- ##### Sample size should not be more than 10% of the population data and it should be atleast 30 sized.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_data = [np.random.randint(10, 100) for i in range(10000)]

len(pop_data)
```

```python
pop_table = pd.DataFrame({
    "Pop_data":pop_data
})

pop_table.head()
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Pop_data", data=pop_table)
plt.show()
```

```python
np.random.choice(pop_data)
```

```python
sam_mean = []
for no_sample in range(60):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sam_mean.append(np.mean(sample_data))
```

```python
sample_M = pd.DataFrame({
    "Sample_mean":sam_mean
})

sample_M.head(4)
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x = "Sample_mean", data = sample_M)
plt.show()
```

```python
np.mean(pop_data), np.mean(sam_mean)
```
