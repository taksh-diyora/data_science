# Percentage Percentile Quatile

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('data/titanic.csv')
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset['Age'].fillna(dataset['Age'].mean(), inplace=True)
```

```python
np.percentile(dataset["Age"], 25)
```

```python
np.percentile(dataset["Age"], 0)
```

```python
dataset["Age"].min()
```

```python
if dataset["Age"].median() == np.percentile(dataset["Age"], 50):
    print("Equal")
else:
    print("Not equal")
```

```python
dataset.describe()
```

```python
sns.boxplot(x="Age", data=dataset)
```
