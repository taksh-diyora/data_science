# Determining Skewness
* ### how much the data is Spread and whether there exist outliers or not

## Frequency & Cumulative Distribution

* ### generally, Histogram is used to determine this

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("data/titanic.csv")
```

```python
dataset.head(3)
```

```python
dataset["Age"].skew() #Positive-> Positive Skewness
```

```python
sns.histplot(x = "Age", data = dataset)
```
