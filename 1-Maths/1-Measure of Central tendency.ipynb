```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('data/titanic.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
mn = dataset["Fare"].mean()
md = dataset["Fare"].median()
mo = dataset["Fare"].mode()[0] #Used indexing as dataset["Fare"].mode() returns a series
```

```python
sns.histplot(x = "Fare", data=dataset, bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0, 300)], [i for i in range(0, 300)], color = "red")
plt.plot([md for i in range(0, 300)], [i for i in range(0, 300)], color = "blue")
plt.plot([mo for i in range(0, 300)], [i for i in range(0, 300)], color = "green")
plt.show()
```
