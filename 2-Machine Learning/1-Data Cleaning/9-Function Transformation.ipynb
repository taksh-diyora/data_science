```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as  np
```

```python
dataset = pd.read_csv('data/loan.csv')
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func = np.log1p)
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_tf"] = ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(dataset["CoapplicantIncome"])
plt.title("Before")
plt.subplot(1,2,2)
sns.distplot(dataset["CoapplicantIncome_tf"])
plt.title("after")
plt.show()
```

# Now without outliers

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)

IQR = q3 - q1
```

```python
min_v = q1 - 1.5*IQR
max_v = q3 + 1.5*IQR
```

```python
dataset = dataset[dataset["CoapplicantIncome"] <= max_v]
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
ft = FunctionTransformer(func = np.log1p)
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_tf"] = ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(dataset["CoapplicantIncome"])
plt.title("Before")
plt.subplot(1,2,2)
sns.distplot(dataset["CoapplicantIncome_tf"])
plt.title("after")
plt.show()
```
