```python
# Handling missing values using scikit-learn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('data/titanic.csv')
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include=["float64", "int64"]).columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy = "mean")
ar = si.fit_transform(dataset[['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])
```

```python
new_dataset = pd.DataFrame(ar, columns=['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare'])
```

```python
new_dataset.isnull().sum()
```

```python
new_dataset.isnull().sum().sum()
```
