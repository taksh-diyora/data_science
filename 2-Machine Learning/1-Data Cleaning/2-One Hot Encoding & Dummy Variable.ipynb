### Get Dummies method using pandas

```python
en_data = dataset[["sex", "smoker"]]
```

```python
pd.get_dummies(en_data).info()
```

### using Scikit learn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop = 'first')
ar = ohe.fit_transform(en_data).toarray()
```
