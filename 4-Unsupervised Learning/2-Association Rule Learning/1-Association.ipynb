### How does association rule learning work: 

- ##### Support
- ##### Confidence
- ##### Lift

## Support:

- ##### The first step for us and the algorithm is to find frequently bought items. It is a straightforward calculation that is based on frequency

- ##### Support(A) = Transactions(A)/Total Transactions

## Confidence:

- ##### This will tell us how confident(based on our data) we can be that an item will be purchased, given that another item has been purchased
- ##### Conditional Probability basically
- ##### P(y/x) = number of transactions containing x and y / number of transactions containing x

## Lift:

- ##### Given that different items are bought at different frequencies (Strong association)

- ##### lift(x=>y) = supp(x intersection y) / (supp(x) * supp(y))

- supp(x intersection y) : number of transactions containing x and y
- supp(x) : number of transactions containing x
- supp(y) : number of transactions containing y


- #### lift = 1: no association
- #### lift < 1: more likely to be bought separately
- #### lift > 1: more likely to be bought together
