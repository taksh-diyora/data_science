## *1. K-means Clustering**

```python
# when we need to find Categorical data then we are use K-means Clustering
```

K-Means Clustering is an Unsupervised Learning algorithm, which groups the unlabeled dataset into different clusters.

K defines the number of pre-defined clusters that need to be created in the process

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib.image as mpimg
```

```python
plt.figure(figsize=(5, 4))
img = mpimg.imread('Image/k-means_group.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
dataset = pd.read_csv('data/iris.csv')
```

```python
dataset.head(5)
```

```python
sns.pairplot(dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2,21):
    km = KMeans(n_clusters=i, init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2, 21)], wcss, marker="o")
plt.grid(axis='both')
plt.xticks([i for i in range(2, 21)])
plt.xlabel("No of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmn = KMeans(n_clusters=3)
dataset['predict'] = kmn.fit_predict(dataset)
```

```python
dataset.head()
```

```python
sns.pairplot(data=dataset, hue="predict")
plt.show()
```

# **2. Hirarchical**

```python
# when we need to find Continous data then we are use Hirarchical Clustering
```

It is used to group the unlabeled datasets into a cluster and also known as hierarchical cluster analysis or HCA.

In this algorithm, we develop the hierarchy of clusters in the form of a tree, and this tree-shaped structure is known as the dendrogram.

```python
plt.figure(figsize=(5, 4))
img = mpimg.imread('Image/Dendrogram.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```

The Dendrogram is a tree-like structure that is mainly used to store each step as a memory that the HC algorithm performs.

The Dendrogram plot, the Y-axis shows the Euclidean distances between the data points, and the x-axis shows all the data points of the given dataset.

## Hirarchical clustering technique has two approaches:
### 1. Agglomerative: Agglomerative is a bottom-up approach, in which the algorithm starts with taking all data points as single clusters and merging them until one cluster is left.

### 2. Divisive: Divisive algorithm is the reverse of the agglomerative algorithm top-down approach.

### **Linkage:**

The closest distance between the two clusters is crucial for the hierarchical clustering.

There are various ways to calculate the distance between two clusters, and these ways decide the rule for clustering.

These measures are called Linkage Methods.

##### 1. Single Linkage
##### 2. Complete Linkage
##### 3. Average Linkage
##### 4. Centroid Linkage

```python
dataset = pd.read_csv('data/Iris.csv')
```

```python
dataset.head(3)
```

```python
sns.pairplot(data = dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(dataset, method='single', metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2, linkage='single')
dataset["predict"] = ac.fit_predict(dataset)
dataset["predict"].head()
```

```python
sns.pairplot(data = dataset, hue = "predict")
plt.show()
```

## Silhouette Score

```python
dataset = pd.read_csv('data/iris.csv')
```

```python
dataset.head(5)
```

```python
sns.pairplot(dataset)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(2,21):
    km = KMeans(n_clusters=i, init='k-means++')
    km.fit(dataset)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2, 21)], wcss, marker="o")
plt.grid(axis='both')
plt.xticks([i for i in range(2, 21)])
plt.xlabel("No of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
kmn = KMeans(n_clusters=3)
dataset['predict'] = kmn.fit_predict(dataset)
```

```python
kmn.labels_
```

```python
silhouette_score(dataset, kmn.labels_)
```

```python
for i in range(2, 21):
    kmn1 = KMeans(n_clusters=i)
    kmn1.fit(dataset)
    print(f"{i}. Silhouette score: {silhouette_score(dataset, kmn1.labels_)}")


## 2 is the best numbers of clusters
```
