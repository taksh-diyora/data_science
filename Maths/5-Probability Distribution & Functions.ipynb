# Random Variables

* ##### Discrete Random Variable: A random variable that takes on a countable number of distinct values. ex: throwing dice
* ##### Continuous Random Variable: A random variable that can take on any value within a given range or interval. ex: Counting heights of 100 people

# Probability

##### P(A) = Number of timems A occurs / Total number of possible outcomes

* ##### Continuous Probability Distribution



### Probability functions:

* ##### Probability Density function(PDF): when we are working on a continuous distribution
* ##### Probability Mass function(PMF): when we are working on a discrete distribution
* ##### Cumulative Density function(CDF): When we are working on a cumulative data

## Normal Distribution function: 

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

* #### also known as Gaussian Distribution


#### 3 STD:
- ##### 1st STD: data coverd in 68.26% by taking mean as central value
- ##### 2st STD: data coverd in 95.44% by taking mean as central value
- ##### 3st STD: data coverd in 99.72% by taking mean as central value

### Standard Normal Distribution - (Centralized curve of Normal Distribution where will be denoted as 0)
