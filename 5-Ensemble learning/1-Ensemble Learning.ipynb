##### The **ensemble methods** in ML combine the insights obtained from multiple learning models to facilitate accurate and improved decisions.


##### 2 main types:

- ##### 1. Bagging (Bootstrap Aggregating)
- ##### 2. Boosting

#### *Basic Ensemble Techniques*:
- ##### Max Voting
- ##### Averaging
- ##### Weighted Average

#### *Advanced Ensemble Techniques*:
- ##### Stacking
- ##### Blending
- ##### Bagging
- ##### Boosting

### *Algorithms based on Baggin and Boosting*

- #### **Bagging Algorithms**:
- ##### Bagging meta-estimator
- ##### Random Forest

- #### **Boosting Algorithms**:
- ##### AdaBoost
- ##### GBM
- ##### XGBM
- ##### Light GBM
- ##### CatBoost
