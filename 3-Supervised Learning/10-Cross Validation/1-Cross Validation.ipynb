## Multiple methods are there for cross validation

cross validation is generally used to get a blur look over the range of highest accuracy of a model before training it properly

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('data/package.csv')
```

```python
dataset.head(4)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["package"]
```

```python
new_data = dataset.head(10)
```

```python
x_new = new_data.iloc[:,:-1]
y_new = new_data["package"]
```

```python
from sklearn.model_selection import LeaveOneOut, LeavePOut, KFold, StratifiedKFold
```

```python
loo = LeaveOneOut()

for train, test in loo.split(x_new, y_new):
    print(train, test)
```

```python
lpo = LeavePOut(p=2)

for train, test in lpo.split(x_new, y_new):
    print(train, test)
```

```python
kf = KFold(n_splits = 5)

for train, test in kf.split(x_new, y_new):
    print(train, test)
```
