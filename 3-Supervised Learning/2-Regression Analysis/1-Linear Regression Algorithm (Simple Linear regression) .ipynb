# Simple Linear Regression :

#### When only 1 independent data series is available

##### y = mx + c

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as  plt
```

```python
dataset = pd.read_csv('data/package.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
inp = dataset[["cgpa"]]
out = dataset["package"]
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x=inp["cgpa"], y=out)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(inp, out, test_size=0.25, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.coef_ #m
```

```python
lr.intercept_ #c
```

y = 0.57111357x - 0.9998951779221357

```python
lr.score(x_train, y_train), lr.score(x_test, y_test)
```

```python
y_pred = lr.predict(inp)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x=inp["cgpa"], y=out)
plt.plot(dataset["cgpa"], y_pred, color="red")
plt.legend(["org_data", "predict line"])
plt.show()
```
