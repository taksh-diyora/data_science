```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('data/salary.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# Age vs Salary

sns.scatterplot(x = "Age", y = "Salary", data = dataset)
plt.show()
```

```python
# YearsExperience vs Salary

sns.scatterplot(x = "YearsExperience", y = "Salary", data = dataset)
plt.show()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
sns.heatmap(data = dataset.corr(), annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
inp = dataset.iloc[:,:-1]
out = dataset["Salary"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(inp, out, train_size=0.5, random_state=42)
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
pred_data = lr.score(inp, out)
```
