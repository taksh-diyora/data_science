## Types of SVMs
### 1. **Simple SVM:** Typically used for linear regression and classification problems.
### 2. **Kernal SVM:** Has more flexibility for non-linear data because you can add more features to fit a hyperplane istead of a two-dimensional space.

## **Kernal Function**

kernel functions play a crucial role in transforming input data into a higher-dimensional space.

The primary purpose of kernel functions is to allow SVMs to handle non-linearly separable data by implicitly mapping the input data into a higher-dimensional feature space where linear separation may be more feasible.

This transformation is done without explicitly calculating the coordinates points in that higher-dimensional space.

⭐ 1. Linear Kernel
Simple Definition

Linear means a straight-line relation like
y = mx + c

In SVM kernel: measuring similarity using simple dot product.

LaTeX
$$
K(x_i, x_j) = x_i^T x_j + b
$$

⭐ 2. Polynomial Kernel
Simple Definition

Polynomial means
m·x + c,
(m·x + c)^2,
(m·x + c)^3, etc.

It allows curved boundaries (2nd degree, 3rd degree…).

LaTeX
$$
K(x_i, x_j) = (\gamma \, x_i^T x_j + b)^{\,N}
$$

⭐ 3. Gaussian RBF Kernel (Radial Basis Function)
Simple Definition

Measures similarity based on distance.

If two points are close → high value

If far → almost 0

Creates circular (radial) decision boundaries.

LaTeX
$$
K(x_i, x_j) = \exp\left(-\gamma \, \|x_i - x_j\|^{2}\right)
$$

⭐ 4. Sigmoid Kernel (Complete Version)
Simple Definition

Works like a neural network activation function
tanh(m·x + c)

Creates smooth S-shaped decision boundaries.

LaTeX
$$
K(x_i, x_j) = \tanh(\alpha \, x_i^T x_j + c)
$$

## Kernel Functions

### 1. Linear Kernel
$$
K(x_i, x_j) = x_i^T x_j + b
$$
Linear kernel means straight-line decision boundary (like \(y = mx + c\)).

---

### 2. Polynomial Kernel
$$
K(x_i, x_j) = (\gamma \, x_i^T x_j + b)^{\,N}
$$
Polynomial means curved boundaries (square, cube, etc).

---

### 3. Gaussian RBF Kernel
$$
K(x_i, x_j) = \exp\left(-\gamma \, \|x_i - x_j\|^{2}\right)
$$
RBF measures similarity based on distance (close = high similarity).

---

### 4. Sigmoid Kernel
$$
K(x_i, x_j) = \tanh(\alpha \, x_i^T x_j + c)
$$
Sigmoid acts like a neural network activation function.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('data/placement.csv')
```

```python
dataset.head()
```

```python
sns.scatterplot(data = dataset, x = "cgpa", y = "score", hue = "placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
```

## Kernal : Linear

```python
sv = SVC(kernel = 'linear')
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)
```

```python
sv.score(x_train, y_train)
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), sv)
plt.show()
```

# Kernal Polynomial

```python
sv = SVC(kernel = 'poly')
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)
```

```python
sv.score(x_train, y_train)
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), sv)
plt.show()
```

# Kernal : RBF

```python
sv = SVC(kernel = 'rbf')
sv.fit(x_train, y_train)
```

```python
sv.score(x_train, y_train)
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), sv)
plt.show()
```

# Kernal : Sigmoid

## what if data is not linearly separable?

```python
dataset = pd.read_csv('data/nonlinear.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x = "x1", y = "x2", data=dataset, hue = "label")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["label"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
```

```python
for kern in kernels:
    sv = SVC(kernel=kern)
    sv.fit(x_train, y_train)
    plot_decision_regions(x.to_numpy(), y.to_numpy(), sv)
    plt.show()
    print(f"{kern}: \n test score: {sv.score(x_test, y_test)} \n train score: {sv.score(x_train, y_train)} \n\n")
```
