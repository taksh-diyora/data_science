### Regression Cost Function


#### Types of error: 

##### 1. MSE(Mean square error) :
- ###### Differenciable eqn 
- ###### Outliers can change the output easily

##### 2. MAE(Mean Absolute error) :
- ###### More robust for outliers
- ###### Not a differenciable method

##### 3. RMSE(Root Mean square error) : (Generally considered as the best)
- ###### Differenciable + robust for outliers
