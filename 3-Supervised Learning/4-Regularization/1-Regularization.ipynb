## Regularization

- #### Improved accuracy of the model
- #### Minimizes overfitting
- #### Feature selection(kind of)

## Main 2 types of Regularization techniques:

- ##### Ridge Regularization /L2
- ##### Lasso Regularization /L1


- ##### Elastic net(combination of Ridge and Lasoo Regularization)

### Lasso Regularization (L1):

- #### changes cost function through which it selects features

### Ridge Regularization (L2):

- #### Extension to linear Regression that introduces a regularization term to reduce model complexity and help prevent overfitting.
